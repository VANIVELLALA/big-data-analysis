```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_excel("food_sales_dataset.csv.xlsx")

# Group by Product and sum sales
product_sales = df.groupby('Product')['Total'].sum().sort_values(ascending=False)

# Plot
plt.figure(figsize=(8, 5))
product_sales.plot(kind='bar')
plt.title("Total Sales by Product")
plt.xlabel("Product")
plt.ylabel("Sales")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
df['Date'] = pd.to_datetime(df['Date'])
df['Month'] = df['Date'].dt.to_period('M').astype(str)

monthly_sales = df.groupby('Month')['Total'].sum()

plt.figure(figsize=(8, 4))
monthly_sales.plot(kind='line', marker='o')
plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Total Sales")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
category_sales = df.groupby('Category')['Total'].sum()

plt.figure(figsize=(6, 6))
category_sales.plot(kind='pie', autopct='%1.1f%%', startangle=140)
plt.title("Sales Distribution by Category")
plt.ylabel("")  # Hide y-label
plt.tight_layout()
plt.show()
```

```python
city_sales = df.groupby('City')['Total'].sum().sort_values(ascending=False).head(5)

plt.figure(figsize=(7, 4))
city_sales.plot(kind='barh', color='skyblue')
plt.title("Top 5 Cities by Sales")
plt.xlabel("Total Sales")
plt.ylabel("City")
plt.tight_layout()
plt.show()
```

```python
product_qty = df.groupby('Product')['Quantity'].sum().sort_values(ascending=False)

plt.figure(figsize=(8, 5))
product_qty.plot(kind='bar', color='orange')
plt.title("Total Quantity Sold by Product")
plt.xlabel("Product")
plt.ylabel("Quantity")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
